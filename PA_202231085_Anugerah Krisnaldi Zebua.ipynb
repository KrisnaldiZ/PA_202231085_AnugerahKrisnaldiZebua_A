```python
import cv2
import matplotlib.pyplot as plt

# Baca gambar dari file
image = cv2.imread('aku ganteng.JPG')

# Ubah warna gambar dari BGR (default OpenCV) ke RGB (default Matplotlib)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Tampilkan gambar menggunakan Matplotlib
plt.imshow(image_rgb)
plt.axis('off')  # Hilangkan sumbu
plt.title('Original Image')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Baca gambar dari file
image = cv2.imread('aku ganteng.JPG', cv2.IMREAD_GRAYSCALE)

# Terapkan Canny edge detection
edges = cv2.Canny(image, 100, 200)

# Tampilkan gambar asli dan hasil edge detection menggunakan Matplotlib
plt.figure(figsize=(10, 5))

# Gambar asli
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Hasil Canny edge detection
plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Detection')
plt.axis('off')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Baca gambar dari file
image = cv2.imread('aku ganteng.JPG')
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Terapkan threshold untuk binarize gambar
_, thresh = cv2.threshold(image_gray, 127, 255, cv2.THRESH_BINARY)

# Temukan kontur
contours, _ = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Gambar kontur pada gambar asli
image_contours = image.copy()
cv2.drawContours(image_contours, contours, -1, (0, 255, 0), 2)

# Ubah gambar asli dan gambar dengan kontur ke format RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_contours_rgb = cv2.cvtColor(image_contours, cv2.COLOR_BGR2RGB)

# Tampilkan gambar asli dan gambar dengan kontur menggunakan Matplotlib
plt.figure(figsize=(10, 5))

# Gambar asli
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

# Gambar dengan kontur
plt.subplot(1, 2, 2)
plt.imshow(image_contours_rgb)
plt.title('Contours Detection')
plt.axis('off')

plt.show()
```
